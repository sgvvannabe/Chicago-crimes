```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the crimes by month dataframe
df_month = pd.read_csv('CSV Dataset/Crimes_month.csv', index_col=[0])
```

```python
df_month
```

```python
df_month.columns
```

```python
# Drop the columns that are classified as Non-criminal
df_month.drop(['NON - CRIMINAL', 'NON-CRIMINAL', 'NON-CRIMINAL (SUBJECT SPECIFIED)'], inplace=True, axis=1)
```

```python
# Create month in int and str
month_int = list(range(1, 13))
month_str = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
```

```python
# Set the x-position for the grouped bar graphs
barwidth = 0.25

bar16 = [x - barwidth for x in month_int]
bar17 = month_int
bar18 = [x + barwidth for x in month_int]
```

```python
# Use 'ggplot' style
plt.style.use('ggplot')
```

```python
# Make a list of series that can be looped for graphing
series_list = []

for col, data in df_month.items():
    if col != 'Month':
        series_list.append(data)
```

```python
# Loop over the series_list and bar graph 
for crime in series_list:
    # Set the figure size
    fig, ax = plt.subplots(figsize=(15,5))

    # Set background color and grid
    ax.set_facecolor('white')
    ax.grid(axis='y', color='black', alpha=0.2)

    # Plot the bar graphs
    plt.bar(x=bar16, height=crime.iloc[0:12], width=barwidth, label='2016')
    plt.bar(x=bar17, height=crime.iloc[12:24], width=barwidth, label='2017')
    plt.bar(x=bar18, height=crime.iloc[24:36], width=barwidth, label='2018')

    # Set the ticks to month in words 
    plt.xticks(month_int, month_str)

    # Set title, labels, and legend
    plt.title(crime.name.capitalize() + ' occurrence by each year', fontsize=20, fontweight='bold')
    plt.xlabel('Month', fontsize=16, fontweight='bold')
    plt.ylabel('Number of crimes', fontsize=16, fontweight='bold')
    plt.legend(fontsize=11, loc='upper right', frameon=True, facecolor='White', edgecolor='black')
    
    # Set ylim higher so graph does not collide with the legend
    plt.ylim(0, max(crime) * 1.25)
    
    plt.show()
```

```python
# Loop over the series_list and line graph 
for crime in series_list:
    # Set the figure size
    fig, ax = plt.subplots(figsize=(15,5))

    # Set background color and grid
    ax.set_facecolor('white')
    ax.grid(axis='x', color='black', alpha=0.2)
    ax.grid(axis='y', alpha=0)

    # Plot the line graphs
    sns.lineplot(x=month_int, y=crime.iloc[0:12], label='2016')
    sns.lineplot(x=month_int, y=crime.iloc[12:24], label='2017')
    sns.lineplot(x=month_int, y=crime.iloc[24:36], label='2018')

    # Set the ticks to month in words 
    plt.xticks(month_int, month_str)

    # Set title, labels, and legend
    plt.title(crime.name.capitalize() + ' occurrence by each year', fontsize=20, fontweight='bold')
    plt.xlabel('Month', fontsize=16, fontweight='bold')
    plt.ylabel('Number of crimes', fontsize=16, fontweight='bold')
    plt.legend(fontsize=13, loc='upper right', frameon=True, facecolor='White', edgecolor='black')
    
    plt.ylim(min(crime) * 0.95, max(crime) * 1.05)
    
    plt.show()
```

```python
# Load the crimes by hour dataframe
df_hour = pd.read_csv('CSV Dataset/Crimes_hour.csv', index_col=[0])
```

```python
# Group the hour dataframe
grouped_df_hour = df_hour.groupby(['Primary Type'])['Hour']
```

```python
# Loop over the groupby dataframe and histogram  
for crime, data in grouped_df_hour:
    # Set the figure size
    fig, ax = plt.subplots(figsize=(8,3))

    # Set background color and grid
    ax.set_facecolor('white')
    ax.grid(axis='x', color='black', alpha=0.2)
    ax.grid(axis='y', alpha=0)

    # Plot the histogram
    sns.distplot(data, bins=24, hist_kws={'alpha':0.5}, kde_kws={'color':'g', 'lw':0.7})
    
    # Set the ticks appropriately 
    plt.xticks(range(0, 25, 3))

    # Set title and labels
    plt.title(crime.capitalize() + ' occurrence by the hour', fontsize=16, fontweight='bold')
    plt.xlabel('Hour', fontsize=14, fontweight='bold')
    plt.ylabel('Frequency', fontsize=14, fontweight='bold')
    plt.xlim(0, 25)
    
    plt.show()
```

```python
# Horizontal bar graph the average hour that each crime occurs
fig, ax = plt.subplots(figsize=(10, 10))

# Set background color to white
ax.set_facecolor('white')
ax.grid(axis='x', color='black', alpha=0.2)

# Plot the graph
df_hour.groupby('Primary Type').Hour.mean().plot(kind='barh', alpha=0.7)

# Set the ticks appropriately 
plt.xticks(range(0, 24, 4))

# Set title and labels
plt.title('Average hour of crime occurence', fontsize=14)
plt.xlabel('Hour', fontsize=13)

plt.show()
```

```python
# Horizontal bar graph the median hour that each crime occurs
fig, ax = plt.subplots(figsize=(10, 10))

# Set background color to white
ax.set_facecolor('white')
ax.grid(axis='x', color='black', alpha=0.2)

# Plot the graph
df_hour.groupby('Primary Type').Hour.describe()['50%'].plot(kind='barh', alpha=0.7)

# Set the ticks appropriately 
plt.xticks(range(0, 24, 4))

# Set title and labels
plt.title('Median hour of crime occurence', fontsize=14)
plt.xlabel('Hour', fontsize=13)

plt.show()
```
