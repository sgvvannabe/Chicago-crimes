# Hypothesis for Shapiro-Wilk test

H0: The sample is a Gaussian distribution.

Ha: The sample is not a Gaussian distribution.

```python
# Shapiro-Wilk test to evaluate if the data was drawn from a Gaussian distribution.
def shapiro_test(col, data, lst, year):
    # Generate a probability plot 
    stats.probplot(data, dist="norm", plot=plt, rvalue=True)
    plt.title(str(year) + ' ' + col.capitalize() + ' Probability Plot')
    plt.show()
    
    # Report stat and p-value
    stat, p = stats.shapiro(data)
    print('Statistics=%.2f, p=%.2f' % (stat, p))
    
    # Test null hypothesis
    alpha = 0.05
    if p > alpha:
        print(col.capitalize(), 'looks Gaussian (fail to reject H0)\n')
        lst.append(0)
    else:
        print(col.capitalize(), 'does not look Gaussian (reject H0)\n')
        lst.append(1)
```

```python
# Test the hypothesis for each crime for every month in the year of 2016
year_2016 = []

for col, data in df_month.loc[2016].items():
    shapiro_test(col, data, year_2016, 2016)
```

```python
# Test the hypothesis for each crime for every month in the year of 2017
year_2017 = []

for col, data in df_month.loc[2017].items():
    shapiro_test(col, data, year_2017, 2017)
```

```python
# Test the hypothesis for each crime for every month in the year of 2018
year_2018 = []

for col, data in df_month.loc[2018].items():
    shapiro_test(col, data, year_2018, 2018)
```

```python
# Create a dummy dataframe to check which columns rejected null hypothesis
_ = pd.DataFrame([year_2016, year_2017, year_2018])
_.columns = list(df_month.columns)

rej_columns = []

for col, data in _.items():
    if sum(data) != 0:
        print(col)
        rej_columns.append(col)
```

The crimes listed above will be tested using Nonparametric methods.

# Hypothesis for one-way ANOVA test
H0: The sample has no statistically significant differences between the group means.

Ha: The sample significant differences between the group means.

```python
# Define a function for ANOVA test
def anova_test(data1, data2, data3, col):
    f, p = stats.f_oneway(data1, data2, data3)
    print('F-statistic = %.2f, p = %.2f' % (f, p))
    alpha = 0.05
    if p > alpha:
        print('Fail to reject H0 for', col.capitalize(), '\n')
    else:
        print('Reject H0 for', col.capitalize(), '\n')
```

```python
# Loop over each column in the dataframe and perform ANOVA test
for col, data in df_month.items():
    if col not in rej_columns:
        anova_test(data[0:12], data[12:24], data[24:36], col)
```

# Hypothesis for Kruskal-Wallis H test

H0: The population median of all of the groups are equal.

Ha: The population median of all of the groups are not equal.

```python
# Recall which nonparamatic columns needed to be tested 
rej_columns
```

```python
# Create a dataframe only including columns which are not normally distributed
df_kruskal = df_month[rej_columns]
```

```python
# Kruskal-Wallis H-test
def kruskal_test(data1, data2, data3, col):  
    stat, p = stats.kruskal(data1, data2, data3)
    print(col.capitalize(), '- H statistics=%.2f, p=%.2f' % (stat, p))
    
    crit_chi = 5.9915
    
    if crit_chi < stat:
        print('Reject null hypothesis\n')
    else:
        print('Failed to reject null hypothesis\n')
```

For 2 degrees of freedom and an alpha level of 0.05, the <u>__critical chi square__</u> value is <u>__5.9915__</u>.

It is more accurate to test for null hypothesis using H statistic and critical chi square value rather than p-value in this test.

If the critical chi-square value is less than the H statistic, reject the null hypothesis that the medians are equal.

If the chi-square value is more than the H statistic, there is not enough evidence to suggest that the medians are unequal.

```python
# Loop over each column in the dataframe and perform ANOVA test
for col, data in df_kruskal.items():
    kruskal_test(data[0:12], data[12:24], data[24:36], col)
```

# Month, hour, and weekday correlations

```python
# Load the crimes by day and hour dataframe
df = pd.read_csv('CSV Dataset/Date_area.csv', index_col=[0], parse_dates=['Date'])
```

```python
df.head()
```

```python
# Define a function which will be needed to generate heatmaps to show correlations
def gen_heatmap(dataframe):
    
    x_hours = list(range(0, 24))
    primtype = np.sort(dataframe['Primary Type'].unique())
    weekdays = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
    
    for crime in primtype:
        _ = dataframe.loc[dataframe['Primary Type'] == crime]
        fig, ax = plt.subplots(1, 3, figsize=(20, 5))
        
        x = pd.pivot_table(_, index='Month', columns='Hour', values='Primary Type', aggfunc='count', fill_value=0)
        map1 = sns.heatmap(x, fmt='.1g', xticklabels=x_hours, ax=ax[0])
        map1.set_xlabel('Hour')
        map1.set_title(crime)
        
        y = pd.pivot_table(_, index='Month', columns='Weekday', values='Primary Type', aggfunc='count', fill_value=0)
        map2 = sns.heatmap(y, fmt='.1g', xticklabels=weekdays, ax=ax[1])
        map2.set_xlabel('Day')
        map2.set_title(crime)
        
        z = pd.pivot_table(_, index='Weekday', columns='Hour', values='Primary Type', aggfunc='count', fill_value=0)
        map3 = sns.heatmap(z, fmt='.1g', xticklabels=x_hours, yticklabels=weekdays, ax=ax[2])
        map3.set_xlabel('Hour')
        map3.set_title(crime)

        plt.show()       
```

```python
df_heatmap = df[['Month', 'Hour', 'Weekday', 'Primary Type']]
gen_heatmap(df_heatmap)
```

From the above heatmap, it is observed that <u>Month vs. Hour</u> and <u>Weekday vs. Hour</u> heatmap are very simliar.

This indicates that the variable 'hour' is very robust in measuring the crimes rates.

Just because 'hour' seems to be the most significant variable, it does not mean the other two variables (Month and Weekday) can be ignored.

For example, looking at Weekday vs. Hour heatmap for the offense involving children, the crime mostly occurred on Friday, Sunday, and Monday. You could think to yourself there's no pattern between the weekdays. But looking at our calendar, 1st of January for 2016, 2017, and 2018 were Friday, Sunday and Monday respectively. By observing the other two heatmaps (Month vs. Hour and Month vs. Day), we can see that this crime mostly occurred  on January 1st, between 12 AM and 1 AM.
