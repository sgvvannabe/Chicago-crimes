```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the crimes by month dataframe
df_month = pd.read_csv('CSV Dataset/Crimes_month.csv', index_col=[0])
```

```python
df_month.head()
```

```python
# Use 'ggplot' style
plt.style.use('ggplot')
```

# Type of crime occurrence by each year

```python
for col, data in df_month.items():
    if col != 'Month':

        fig, ax = plt.subplots(figsize=(6,5))

        # Set background color and grid
        ax.set_facecolor('white')
        ax.grid(axis='y', color='black', alpha=0.2)

        # Plot the bar graphs
        plt.bar(x=1, height=data[0:12], label='2016')
        plt.bar(x=2, height=data[12:24], label='2017')
        plt.bar(x=3, height=data[24:36], label='2018')

        # Set the ticks to month in words 
        plt.xticks([1, 2, 3], ['2016', '2017', '2018'])

        # Set title, labels, and legend
        plt.title(col.capitalize() + ' occurrence by each year', fontsize=16)
        plt.xlabel('Year', fontsize=14)
        plt.ylabel('Total count of crimes', fontsize=14)
        plt.legend(fontsize=9, loc='upper right', frameon=True, facecolor='White', edgecolor='black', ncol=3)

        # Set ylim higher so graph does not collide with the legend
        plt.ylim(0, max(data) * 1.2)
        
        plt.show()
```

# Type of crime occurrence by each month and year

```python
# Create month in int and str
month_int = list(range(1, 13))
month_str = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
```

```python
# Make a list of series that can be looped for graphing
series_list = []

for col, data in df_month.items():
    if col != 'Month':
        series_list.append(data)
```

```python
# Loop over the series_list and line graph 
for crime in series_list:
    # Set the figure size
    fig, ax = plt.subplots(figsize=(15,5))

    # Set background color and grid
    ax.set_facecolor('white')
    ax.grid(axis='x', color='black', alpha=0.2)
    ax.grid(axis='y', alpha=0)

    # Plot the line graphs
    sns.lineplot(x=month_int, y=crime.iloc[0:12], label='2016')
    sns.lineplot(x=month_int, y=crime.iloc[12:24], label='2017')
    sns.lineplot(x=month_int, y=crime.iloc[24:36], label='2018', color='black')

    # Set the ticks to month in words 
    plt.xticks(month_int, month_str)

    # Set title, labels, and legend
    plt.title(crime.name.capitalize() + ' occurrence by each month', fontsize=16, fontweight='bold')
    plt.xlabel('Month', fontsize=14)
    plt.ylabel('Number of crimes', fontsize=14)
    plt.legend(fontsize=13, loc='upper right', frameon=True, facecolor='White', edgecolor='black')
    
    plt.ylim(min(crime) * 0.95, max(crime) * 1.1)
    
    plt.show()
```

From visualizing the above line (time-series) graphs, we can see the following trend for different crimes over the months between each year.

# Type of crime occurrence by the hour

```python
# Load the crimes by day and hour dataframe
df_date_area = pd.read_csv('CSV Dataset/Date_area.csv', index_col=[0], parse_dates=['Date'])
```

```python
# Convert 'Hour' column to include minutes as decimals
df_date_area['Hour'] = df_date_area['Date'].dt.hour + (df_date_area['Date'].dt.minute / 60)

# Convert 'Year' to include hours and minutes
df_date_area['Year'] = df_date_area['Date'].dt.year + (df_date_area['Date'].dt.dayofyear / 365.25) + (df_date_area['Hour'] / 24 / 365.25)
```

```python
# Group the dataframe by 'Primary type'
grouped_df_hour = df_date_area.groupby(['Primary Type'])['Hour']
```

```python
# Loop over the groupby dataframe and visualize histogram  
for crime, data in grouped_df_hour:
    # Set the figure size
    fig, ax = plt.subplots(figsize=(15,5))

    # Set background color and grid
    ax.set_facecolor('white')
    ax.grid(axis='x', color='black', alpha=0.2)
    ax.grid(axis='y', alpha=0)

    # Plot the histogram
    plt.hist(data, bins=24, alpha=0.6)
    
    # Set the ticks appropriately 
    plt.xticks(range(0, 25, 3))

    # Set title and labels
    plt.title(crime.capitalize() + ' occurrence by the hour', fontsize=16, fontweight='bold')
    plt.xlabel('Hour', fontsize=14)
    plt.ylabel('Count of crimes', fontsize=14)
    plt.xlim(0, 24)
    
    plt.show()
```

From visualizing the above histogram, we can see the following trends and/or interesting data:
