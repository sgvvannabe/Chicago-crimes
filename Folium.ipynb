```python
import pandas as pd
import numpy as np
import folium
from folium.plugins import HeatMap
```

```python
df_a = pd.read_csv('CSV Dataset/Crimes_comb.csv')
```

```python
df_a.head()
```

```python
df = df_a[['Year', 'Primary Type', 'District', 'Ward', 'Community Area', 'Latitude', 'Longitude']]
```

```python
df.head()
```

```python
a = df[(df['Year'] == 2016) & (df['Primary Type'] == 'THEFT')]
```

```python
a
```

```python
a.dropna(axis=0, inplace=True)
```

```python
a['count'] = 1
```

```python
a
```

```python
def gen_map(chicago=[41.8781, -87.6298], zoom=9):
    chicago_map = folium.Map(location=chicago, controL_scale=True, zoom_start=zoom)
    return chicago_map
```

```python
chicago_map = gen_map()
chicago_map
```

```python
chicago_heatmap = gen_map()
HeatMap(data=a[['Latitude', 'Longitude', 'count']].groupby(['Latitude', 'Longitude']).sum().reset_index().values.tolist(), radius=8).add_to(chicago_heatmap)
chicago_heatmap
```
